## Useful Functions

```python
def detect_logos(path):
    """Detects logos in the file."""

    client = vision.ImageAnnotatorClient(credentials=credentials)

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision.types.Image(content=content)

    response = client.logo_detection(image=image)
    logos = response.logo_annotations
    print('Logos:')
    
    if response.error.message:
        raise Exception(
            '{}\nFor more info on error messages, check: '
            'https://cloud.google.com/apis/design/errors'.format(
                response.error.message))
        
    return logos


def draw_bounding_box(img, vertices, label):
    
    """ Draws a bounding box around the detected logo"""
    
    #v1, v2, v3, v4 = vertices

    # Creates a rectangle upon the coordinates in the annotated data of the bbox
    #cv2.rectangle(img,(v1.x,v1.y),(v3.x,v3.y),(0,255,0),2)
    #cv2.putText(img, label, (v3.x+5,v3.y+5), cv2.FONT_HERSHEY_DUPLEX, 0.4, (0, 255, 0), 1)

    x1, y1, x2, y2, x3, y3, x4, y4 = vertices

    cv2.rectangle(img,(x1,y1),(x3,y3),(0,255,0),2)
    cv2.putText(img, label, (x3+5,y3+5), cv2.FONT_HERSHEY_DUPLEX, 0.4, (0, 255, 0), 1)
    
    
def write_frame(count, frame, pathOut):

    cv2.imwrite(pathOut + 'frame%d.jpg' % count, frame)

    if not cv2.imwrite(pathOut + 'frame%d.jpg' % count, frame):
        raise Exception("Could not write image")


# def load_videos(video_file):
    
#     capture = cv2.VideoCapture(video_file)

#     fps = capture.get(cv2.CAP_PROP_FPS)
    
#     read_flag, frame = capture.read()
    
#     vid_frames = []

#     while (read_flag):
        
#         vid_frames.append(frame)
        
#         read_flag, frame = capture.read()
        
#     vid_frames = np.asarray(vid_frames, dtype='uint8')
    
#     capture.release()
    
#     return vid_frames, fps


def load_videos(video_file):
    
    cap = cv2.VideoCapture(video_file)
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    if (capture.isOpened()== False): 
        print("Error opening video stream or file")
        
        
    count = 0

    while(cap.isOpened()):

        vid_frames = []
        
        #Capture frame-by-frame
        ret, frame = cap.read()

        if ret==True:

            #write_frame(count, frame, pathOut)
            vid_frames.append(frame)
            
            ret, frame = cap.read()
            count += 1
            wait = cv2.waitKey(10)

            if wait==27 & 0xFF == ord('q'):
                break   

        else:
            break

    cap.release()
    cv2.destroyAllWindows()
    
    vid_frames = np.asarray(vid_frames, dtype='uint8')
    
    return vid_frames, fps


def video_to_frames(vid_input, pathOut, url):

    #start the video
    if url == True:
        vPafy = pafy.new(vid_input)
        play = vPafy.getbest(preftype="webm")
        cap = cv2.VideoCapture(play.url)

    else:
        cap = cv2.VideoCapture(vid_input)

    if (cap.isOpened()== False): 
        print("Error opening video stream or file")

    count = 0

    while(cap.isOpened()):

      # Capture frame-by-frame
        ret, frame = cap.read()

        if ret==True:

            write_frame(count, frame, pathOut)

            ret, frame = cap.read()
            count += 1
            wait = cv2.waitKey(10)

            if wait==27 & 0xFF == ord('q'):
                break   

        else:
            break

    cap.release()
    cv2.destroyAllWindows()


def process_frames(pathIn):

    list_logos = dict()
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
 
    #for sorting the file names properly
    files.sort(key = lambda x: int(x[5:-4]))
 
    for i in tqdm_notebook(range(len(files))):

        filename = pathIn + files[i]

        logos = detect_logos(filename)

        list_logos[i] = logos

    return list_logos

def logos_to_pandas(list_logos):
    
    DfLogos = pd.DataFrame([])

    for frame in range(len(list_logos)):
        for logo in range(len(list_logos[frame])):
            DfTmp = pd.DataFrame({'frame':frame, 
                                  'logo':list_logos[frame][logo].description, 
                                  'score':list_logos[frame][logo].score,
                                  'x1':list_logos[frame][logo].bounding_poly.vertices[0].x, 
                                  'y1':list_logos[frame][logo].bounding_poly.vertices[0].y,
                                  'x2':list_logos[frame][logo].bounding_poly.vertices[1].x, 
                                  'y2':list_logos[frame][logo].bounding_poly.vertices[1].y,
                                  'x3':list_logos[frame][logo].bounding_poly.vertices[2].x, 
                                  'y3':list_logos[frame][logo].bounding_poly.vertices[2].y,
                                  'x4':list_logos[frame][logo].bounding_poly.vertices[3].x, 
                                  'y4':list_logos[frame][logo].bounding_poly.vertices[3].y,
                                  'mid':list_logos[frame][logo].mid}, index=[frame])
            DfLogos = DfLogos.append(DfTmp, ignore_index=True)
            
    return DfLogos


def print_logos(pathIn, pathOut, path_annotations):

    img = cv2.imread(filename)

    for logo in logos:

        draw_bounding_box(img, logo.bounding_poly.vertices, logo.description)

    write_frame(count, img, pathOut)

def frames_to_video(vid_frames , pathOut, annotations, fps, bounding=True):

#     frame_array = []
#     files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
 
#     #for sorting the file names properly
#     files.sort(key = lambda x: int(x[5:-4]))
 
    
#     for i in tqdm_notebook(range(len(files))):

#         filename=pathIn + files[i]

#         #reading each files
#         img = cv2.imread(filename)
#         cv2.putText(img, '@BaaschLopez', (1500,1000), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 255), 1)
#         height, width, layers = img.shape
#         size = (width,height)

#         #inserting the frames into an image array
#         frame_array.append(img)
 
    frame_array = []
    
    num_frames, height, width, layers = vid_frames.shape
    
    for i in tqdm_notebook(range(num_frames)):

        #reading each files
        
        #img = cv2.imread(filename)
        img = vid_frames[i]
        cv2.putText(img, '@BaaschLopez', (1500,1000), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 255), 1)

        #inserting the frames into an image array
        frame_array.append(img)
    
    out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, (width, height))
 
    for i in tqdm_notebook(range(num_frames)):
        
        frame_tmp = frame_array[i]
        
        if bounding:
            for index,row in annotations[annotations.frame==i].iterrows():

                label = row['logo']
                vertices = row[['x1','y1','x2','y2','x3','y3','x4','y4']]
                draw_bounding_box(frame_tmp, vertices, label)

        # writing to a image array
        out.write(frame_tmp)
    out.release()

def generate_anno_heatmap(pathIn_Frames, pathOut_annotations_heatmap, DfLogos,  query_logo):

    LogoFrame = DfTopLogos[DfTopLogos.logo == query_logo]
    del LogoFrame['mid']
    del LogoFrame['logo']
    del LogoFrame['score']

    LogoFrame['C'] = LogoFrame.groupby(['frame']).cumcount()

    heatmap_anno = pd.DataFrame([])

    for frame in list(LogoFrame['frame'].unique()):
        dfTmp = pd.DataFrame({'frame': [frame]})
        dfTmp['Count'] = max(LogoFrame[LogoFrame.frame == frame]['C']) + 1

        for index, row in LogoFrame[LogoFrame.frame == frame].iterrows():
            dfTmp_2 = pd.DataFrame(row[['frame','x1','y1','x2','y2','x3','y3','x4','y4']]).T
            dfTmp_2 = dfTmp_2.rename(columns={"x1": "x1%d" % row['C'], 
                                            "y1": "y1%d" % row['C'],
                                            "x2": "x2%d" % row['C'],
                                            "y2": "y2%d" % row['C'],
                                            "x3": "x3%d" % row['C'],
                                            "y3": "y3%d" % row['C'],
                                            "x4": "x4%d" % row['C'],
                                            "y4": "y4%d" % row['C']})

            dfTmp = dfTmp.merge(dfTmp_2, on=['frame'], how='left', sort=False)

        heatmap_anno = heatmap_anno.append(dfTmp.reset_index(), sort=False)

    heatmap_anno['filepath'] = heatmap_anno['frame'].map(lambda x: pathIn_Frames + 'frame{}.jpg'.format(x))
    del heatmap_anno['index']
    heatmap_anno['frame'] = heatmap_anno['filepath']
    del heatmap_anno['filepath']

    heatmap_anno.to_csv(pathOut_annotations_heatmap, header=None, index=None, sep=' ', mode='w')
    
    with open(pathOut_annotations_heatmap, 'r') as f:
        lines = f.readlines()

    # remove spaces
    lines = [line.replace('"', '') for line in lines]
    lines = [line.replace('  ', '') for line in lines]

    # finally, write lines in the file
    with open(pathOut_annotations_heatmap, 'w') as f:
        f.writelines(lines)


def annotate_video(input_video):
    """
    Performs asynchronous video annotation for logo recognition on a local file.

    Args:
      local_file_path Path to local video file, e.g. /path/video.mp4
    """
    #client = videointelligence_v1p3beta1.VideoIntelligenceServiceClient(credentials=credentials)
    client = videointelligence.VideoIntelligenceServiceClient(credentials=credentials)

    # local_file_path = 'resources/googlework_short.mp4'
    with io.open(input_video, 'rb') as f:
        input_content = f.read()
        
    features = [enums.Feature.LOGO_RECOGNITION]

    operation = client.annotate_video(input_content=input_content, features=features)
    print('\nProcessing video for logo annotations:')
    
    print(u'Waiting for operation to complete...')
    response = operation.result(timeout=270)
    print('\nFinished processing.')
    
    return response

def anno_to_pandas(pathIn_Video, result):
    
    vcap = cv2.VideoCapture(pathIn_Video) # 0=camera

    fps = vcap.get(cv2.CAP_PROP_FPS)
    width, height = (int(vcap.get(cv2.CAP_PROP_FRAME_WIDTH)),
            int(vcap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    annotation_result = result.annotation_results[0]
    
    DfLogos = pd.DataFrame()

    for label in annotation_result.logo_recognition_annotations:
        for track in label.tracks:
            for idx, anno in enumerate(track.timestamped_objects):
                df_tmp = pd.DataFrame({'seconds':anno.time_offset.seconds,
                                       'nanos':anno.time_offset.nanos,
                                       'frame': int((anno.time_offset.seconds + anno.time_offset.nanos/1000000000)*fps)+1,
                                       'logo': label.entity.description,
                                       'x1': int(anno.normalized_bounding_box.left*width), 
                                       'y1': int(anno.normalized_bounding_box.top*height),
                                       'x2': int(anno.normalized_bounding_box.right*width),
                                       'y2': int(anno.normalized_bounding_box.top*height),
                                       'x3': int(anno.normalized_bounding_box.right*width),
                                       'y3': int(anno.normalized_bounding_box.bottom*height),
                                       'x4': int(anno.normalized_bounding_box.left*width),
                                       'y4': int(anno.normalized_bounding_box.bottom*height)}, index=[idx])
                DfLogos = DfLogos.append(df_tmp, sort=False, ignore_index=True)

    return DfLogos

def download_video(video_url, video_title):
    """
    Download a video using youtube url and video title
    """

    ydl_opts = {'outtmpl': pathIn_Video}
    
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([video_url])
```

```python
tmp = DfLogos.groupby(['seconds','logo']).size().reset_index()
tmp['cumul'] = tmp.groupby('logo')[0].cumsum()
```

```python
tmp_dict = dict()

for index, row in tmp.iterrows():
    tmp_dict[row['seconds'],row['logo']] = row['cumul']

tmp_dict = sorted(tmp_dict.items())
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
for logo in list(top_logos.keys()):
    ax.plot([k[0][0] for k in tmp_dict if k[0][1] == logo],[k[1] for k in tmp_dict if k[0][1] == logo])

plt.title('Cumulative Exposure')
plt.xlabel('Second')
plt.ylabel('Freq')
plt.legend(list(top_logos.keys()))
```
